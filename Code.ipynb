```python
ped = pd.read_csv(r"data/akl_ped-2024.csv") # Loading data
# ped = pd.read_csv(r"data\akl_ped-2025.csv") # Loading data
geo_location = pd.read_csv("data/akl_ped-Geodata.csv")

geo_location = geo_location.dropna(subset=["Latitude", "Longitude"]).copy()

print(f"We have coordinates for {len(geo_location)} sensors")
ped.head()
```

```python
ped.rename(columns={"59 High Stret":"59 High Street"}, inplace=True) # Correcting street name

ped = ped.dropna(subset=["Date", "Time"]) # Removing NA
ped = ped[ped["Date"] != "Daylight Savings"] # Removing daylight saving data entry
```

```python
sensor_cols = ped.columns[2:]
sensor_cols
```

```python
ped["Datetime"] = pd.to_datetime(ped["Date"] + " " + ped["Time"].str.split("-").str[0]) # Merging date and time

ped.info()
```

```python
ped.head()
```

```python
ped["Hour"] = ped["Time"].str.split(":").str[0].astype(int)
ped["Month"] = ped["Datetime"].dt.month
ped["DayOfWeek"] = ped["Datetime"].dt.dayofweek
ped["Weekend"] = ped["DayOfWeek"]>=5
ped["DayName"] = ped["Datetime"].dt.day_name()
```

```python
def season_mapping(month):
    """
    Maps month to season based on https://www.newzealand.com/nz/seasons-in-new-zealand/.
    Returns from one of four seasons.
    """
    if month in [3, 4, 5]:
        return "Autumn"
    elif month in [6, 7, 8]:
        return "Winter"
    elif month in [9, 10, 11]:
        return "Spring"
    else:
        return "Summer"
```

```python
ped["Season"] = ped["Month"].apply(season_mapping)
ped
```

```python
ped_long = pd.melt(
    ped,
    id_vars=["Datetime", "Month", "DayOfWeek", "Hour", "Weekend", "DayName","Season"],
    value_vars=sensor_cols.tolist(),
    var_name="location",
    value_name="count"
)
ped_long
```

```python
weather_df = pd.DataFrame(data['hourly'])

weather_df['time'] = pd.to_datetime(weather_df['time'])
weather_df.rename(columns={"time":"Datetime"}, inplace=True) # Updating column name
weather_df.head()
```

```python
merged = pd.merge(ped_long, weather_df, on="Datetime", how="inner")
merged
```

```python
def temperature_mapping(temp):
    """
    Categorizes temperature based on typical Auckland 
    pedestrian comfort levels.
    """
    if temp < 13:
        return "1. Cold (<13°C)"
    elif temp < 17:
        return "2. Cool (13-17°C)"
    elif temp < 22:
        return "3. Mild (17-22°C)"
    else:
        return "4. Warm (>22°C)"

def weather_code_mapping(code):
    """
    Maps weather code based on WMO(World Meteorological Organization) 4677.
    Given the diversity of code. Weather code has been groupped where similar. 
    Returns from one of six weather condition.
    """
    if code <= 3:
        return "1. Clear/Dry"
    elif (4 <= code <= 19) or (30 <= code <= 49):
        return "2. Low Visibility (Fog/Haze)"
    elif code in [20, 21, 50, 51, 58, 60, 61, 80]:
        return "3. Light Rain/Drizzle"
    elif code in [62, 63, 64, 65, 81, 82, 91, 92]:
        return "4. Heavy Rain"
    elif code in [17, 18, 29, 95, 96, 97, 99]:
        return "5. Storm/Thunderstorm"
    else:
        return "6. Other"

def wind_speed_mapping(speed):
    """
    Categorizes wind speed based on the Beaufort Scale 
    and human experience in an urban environment.
    """
    if speed < 12:
        return "1. Light (0-11)"
    elif speed < 29:
        return "2. Moderate (12-28)"
    elif speed < 50:
        return "3. Strong (29-49)" 
    elif speed < 75:
        return "4. Gale (50-74)"
    else:
        return "5. Storm (75+)"
```

```python
merged["temp"] = merged["temperature_2m"].apply(temperature_mapping)
merged["weather"] = merged["weather_code"].apply(weather_code_mapping)
merged["wind"] = merged["wind_speed_10m"].apply(wind_speed_mapping)
merged
```

```python
features = ['count', 'wind_speed_10m', 'rain', 'Hour',"temperature_2m", "weather_code"]
correlation_matrix = merged[features].corr()

count_correlations = correlation_matrix['count'].sort_values(ascending=False)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
working_hour_df = merged[(merged["Hour"] >= 8) & (merged["Hour"] <= 18)]
working_hour_mean = working_hour_df.groupby("DayOfWeek")["count"].mean()

non_working_hour_df = merged[~((merged["Hour"] >= 8) & (merged["Hour"] <= 18))]
non_working_hour_mean = non_working_hour_df.groupby("DayOfWeek")["count"].mean()
```

```python
day_labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

fig, ax = plt.subplots(figsize=(10,5))

ax.plot(working_hour_mean.index, working_hour_mean, color='b', label='Working Hours')
ax.plot(non_working_hour_mean.index, non_working_hour_mean, color='aqua', label='Non-Working Hours')

ax.set_title("Working Hour VS Non-Working Hours Average Pedestrain Count")
ax.set_ylabel("Pedestrain count")
ax.legend()

ax.set_xticks(range(7)) 
ax.set_xticklabels(day_labels)

plt.show()
```

```python
def display_by_season(df, column, title):
    day_labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    fig, axes = plt.subplots(2,2, figsize=(15,10))
    seasons = ["Spring", "Summer", "Autumn", "Winter"]
    variables = sorted(df[column].unique())
    for row in axes:
        for ax in row:
            season = seasons.pop(0)
            season_df = df[df["Season"] == season]
            for i in variables:
                i_df = season_df[season_df[column] == i]
                i_mean = i_df.groupby("DayOfWeek")["count"].mean()
                ax.plot(i_mean.index, i_mean, label=i)
                ax.set_title(f"{season} {title}")
            ax.set_ylabel("Pedestrain count")
            ax.legend(loc='lower left')
            ax.set_xticks(range(7)) 
            ax.set_xticklabels(day_labels)
    plt.tight_layout()
    plt.show()
```
